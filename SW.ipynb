# CLASSIFICATION OF ELCTRONICS BASED ON MODEL NAME

Reading the files from Directory

```python
import os
from operator import itemgetter
files=[]
files=os.listdir("C:/Users/HEMANTH/Desktop/sw-data-set/")
files.sort()
mainDict=[]

def makeDict():
    for file in files:
        d = {}
        with open("sw-data-set/"+file,encoding="utf-8") as f:
            for line in f:
                try:
                    (key, val) = line.split()
                    d[key] = val
                except:
                    pass
        mainDict.append(d)
```

Displaying Models Names

```python
makeDict()
for i in range(len(mainDict)):
      try:   
             print("model Name :"+mainDict[i]['model-name'])
      except: pass
```
